```python
# Visualize the AFI distribution of candidate sequences produced by different design methods

import matplotlib.pyplot as plt
import seaborn as sns

afi_data = []
original_afi = pred['AFI'][0]

# Length-increasing methods
for method, preds in zip(['insert', 'augment', 'duplicate'], [insert_predictions, augment_predictions, duplicate_predictions]):
    for afi in preds['AFI']:
        afi_data.append({'method': method, 'afi': afi, 'group': 'Length-Increasing'})

# Length-preserving methods
for method, preds in zip(['single_point_mutate', 'swap', 'shift'],
                         [single_point_mutation_predictions, swap_predictions, shift_predictions]):
    for afi in preds['AFI']:
        afi_data.append({'method': method, 'afi': afi, 'group': 'Length-Preserving'})

# Length-decreasing methods
for method, preds in zip(['segment', 'delete'], [segment_predictions, delete_predictions]):
    for afi in preds['AFI']:
        afi_data.append({'method': method, 'afi': afi, 'group': 'Length-Decreasing'})

afi_df = pd.DataFrame(afi_data)
plt.figure(figsize=(14, 8))
sns.violinplot(x='method', y='afi', hue='group', data=afi_df, split=True, palette="muted")
plt.axhline(y=original_afi, color='r', linestyle='--', label='Original Sequence AFI')
plt.title('Distribution of AFI values of candidate sequences produced by different design methods')
plt.xlabel('Design Method')
plt.ylabel('Antifungal Index (AFI)')
plt.legend(title='Group', loc='upper right')
plt.show()
```
